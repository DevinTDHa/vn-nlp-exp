sqlite 3 schema:
```sql
CREATE TABLE IF NOT EXISTS "subz" (
"num"INTEGER,
"name"TEXT,
"file"BLOB,
PRIMARY KEY("num")
);

```

Query:

```sql
SELECT num, name 
FROM subz 
WHERE name LIKE '%.vie._cd%' LIMIT 5;
```

````python
import sqlite3


# Create a connection to the SQLite database
db_file = "/media/ducha/SSDSHARED/subs_dump/opensubtitles.org.Actually.Open.Edition.2022.07.25/opensubs.db"

# Connect to the database
conn = sqlite3.connect(db_file)
cursor = conn.cursor()

# Retrieve the BLOB data and file name
query = """SELECT *
FROM subz 
WHERE name LIKE '%.vie.%cd%';
"""
cursor.execute(query)

output_folder = "/media/ducha/SSDSHARED/subs_dump/viet_subs/"
for row in cursor:
    num = row[0]
    file_name = row[1].replace('"', "").split("=", 1)[-1]
    blob_data = row[2]

    print(num, file_name)

    # Write the BLOB data to a file
    try:
        with open(output_folder + file_name, "wb") as f:
            f.write(blob_data)
    except Exception as e:
        print(e)

# Close the connection
conn.close()
````
