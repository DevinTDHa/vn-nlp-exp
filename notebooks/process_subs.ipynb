```python
import pandas as pd
import os
import zipfile

subs_folder = "/media/ducha/SSDSHARED/VN/subs_dump/viet_subs_raw/viet_subs"

def get_sub_list(subs_folder):
    all_files = os.listdir(subs_folder)

    subs_list = set()

    for file_name in all_files:
        file_split = file_name.rsplit(".", 4)
        if len(file_split) != 5:
            continue
    
        base_name, lang, _, _, _ = file_split

        if lang != "vie":
            continue
        if base_name not in subs_list:
            subs_list.add(file_name)


    return subs_list

subs_list = get_sub_list(subs_folder)

print(len(subs_list))
```

## Processing

```python
import regex

output_folder = "/media/ducha/SSDSHARED/VN/subs_dump/viet_subs_processed2"
os.makedirs(output_folder, exist_ok=True)
# Treat files in output_folder as cache
output_files = set(os.listdir(output_folder))


def check_if_file_exists_in_output_folder(file_name):
    return file_name in output_files


def clean_src(src):
    return regex.sub(r"[^\p{L}\p{N}\p{P}\p{Z}]", "", src)


def clean_sentences(sentences):
    return [sentence.strip(" -\n\r\t") for sentence in sentences]
```

```python
from wtpsplit import WtP

wtp = WtP("wtp-bert-mini").to("cuda")
```

```python
from tqdm.notebook import tqdm


# Read all zip files and process them
def process_zip_file(zip_file):
    csv_file_name = os.path.splitext(os.path.basename(zip_file))[0] + ".csv"
    csv_path = os.path.join(output_folder, csv_file_name)

    if check_if_file_exists_in_output_folder(csv_file_name):
        print("File already processed: ", csv_file_name)
        return None

    with zipfile.ZipFile(zip_file, "r") as zip_ref:
        srt_name = find_srt_name(zip_ref)
        if srt_name:
            with zip_ref.open(srt_name) as srt_file:
                contents = srt_file.read().decode("utf-8", errors="ignore")
                line_endings = find_line_endings(contents)
                contents_parsed = [
                    line.split(line_endings[:2])[:3]
                    for line in contents.split(line_endings)
                ]
        else:
            print("SRT file not found in the zip file.")
            return None
    content_df = pd.DataFrame(contents_parsed, columns=["line", "time", "text"])
    content_text_only = remove_html_tags_from_df(content_df)["text"].str.cat(sep=" ")
    splits = wtp.split(content_text_only, lang_code="vi")
    splits_clean = clean_sentences(splits)

    # write to csv file with the same name
    with open(csv_path, "w") as f:
        f.write("\n".join(splits_clean))
    print("processed file: ", csv_path)
```

```python
for file in tqdm(subs_list):
    try:
        process_zip_file(os.path.join(subs_folder, file))
    except Exception as e:
        print("Error processing file: ", file)
        print(e)
```
