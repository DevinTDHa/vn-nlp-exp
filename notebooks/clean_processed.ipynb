```python
all_files = os.listdir(output_folder)

base_names = dict()

for file_name in all_files:
    base_name = file_name.rsplit(".", 4)[0]
    if base_name not in base_names:
        base_names[base_name] = []
    base_names[base_name].append(file_name)
```

```python
for base_name, file_names in base_names.items():
    if len(file_names) > 1:
        print(base_name, len(file_names))
        for file_name in file_names[1:]:
            os.remove(os.path.join(output_folder, file_name))
            print(f"  Removed duplicate {file_name}")
    if file_names[0].rsplit(".", 4)[1] != "vie":
        os.remove(os.path.join(output_folder, file_names[0]))
        print(f"  Removed non-vietnamese {file_names[0]}")
```
